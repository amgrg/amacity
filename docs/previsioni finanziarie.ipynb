```python
import matplotlib.pyplot as plt

# Dati per gli anni, numero di negozi e entrate stimate
anni = ['Primo anno', 'Secondo anno', 'Terzo anno']
negozi = [500, 2000, 5000]
entrate = [300000, 1200000, 3000000]

# Creazione della figura e degli assi
fig, ax1 = plt.subplots()

# Grafico a barre per il numero di negozi
color = 'tab:blue'
ax1.set_xlabel('Anno')
ax1.set_ylabel('Numero di Negozi', color=color)
ax1.bar(anni, negozi, color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)

# Creazione del secondo asse per le entrate
ax2 = ax1.twinx()  
color = 'tab:green'
ax2.set_ylabel('Entrate (â‚¬)', color=color)
ax2.plot(anni, entrate, color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# Titolo del grafico
plt.title('Previsioni Finanziarie: Negozi e Entrate nei Prossimi 3 Anni')

# Mostra il grafico
plt.show()
```
